**6) Create train.txt file**

```
import glob
images_jpg = glob.glob("data/obj/images/*.jpg")
images_png = glob.glob("data/obj/images/*.png")
images_list=images_jpg+images_png
with open("data/train.txt", "w") as f:
    f.write("\n".join(images_list))
```
